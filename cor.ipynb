```python
import os

import pandas as pd

test_data = pd.read_csv('dataset/SWaT/swat2.csv')
print(test_data.columns[:5])
print(test_data.columns[5:16])
print(test_data.columns[16:25])
print(test_data.columns[25:34])
print(test_data.columns[34:47])
print(test_data.columns[47:])
```

```python
import numpy as np

data = test_data.values
cor = {}

for i in range(data.shape[1] - 1):
    a = np.corrcoef(data[:, i], data[:, -1])[0,1]
    if not np.isnan(a):
        cor[test_data.columns[i]] =  a

sorted_cor = sorted(cor.items(), key=lambda x: abs(x[1]), reverse=True)
for i in range(15):
    print(sorted_cor[i] )
```
